# Preliminary Regression Analysis
During our hypothesis testing, we found that there was a signifcant difference between premium charges of smokers. We also found that same amount of premium were charged on both genders. We also look at the difference in premium charges across regions and found that the premium charges on southeast region was significantly different than the northeast region. In this notebook, we will quantify the effects of each predictors with the premium charge with the use of linear regression analysis.

```python
from sklearn import set_config
set_config(transform_output = "pandas")
```

## Data Loading

```python
import numpy as np
import pandas as pd
```

```python
DATASET_URI = "dataset/insurance.csv"
df0 = pd.read_csv(DATASET_URI)

df0.sample(n=10)
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split

# Split the data into feature matrix and target vector
X = df0.iloc[:, :-1]
y = df0.iloc[:, -1]

# Split the data into training, testing set, and evaluation set.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=22, test_size=0.2)
X_test, X_eval, y_test, y_eval = train_test_split(X, y, random_state=22, test_size=0.5)
```

## Building the ML Pipeline

```python
from sklearn.compose import ColumnTransformer, TransformedTargetRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, FunctionTransformer
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector

# Define the ML pipeline
ml_pipeline = make_pipeline(
    # Conduct feature matrix preprocessing
    ColumnTransformer(
        [
            ("scaler", StandardScaler(), ["age", "bmi"]),
            ("passthrough", "passthrough", ["children"]),
            (
                "onehotencoder",
                OneHotEncoder(sparse_output=False, drop="first"),
                ["sex", "smoker", "region"]
            )
        ]
    ),
    
    # Conduct forward feature selection
    SequentialFeatureSelector(
        estimator=LinearRegression(),
        n_features_to_select="auto",
        tol=0.01,
        scoring="neg_mean_squared_error"
    ),
    
    # Conduct target vector preprocessing then regression modelling
    TransformedTargetRegressor(
        regressor=LinearRegression(),
        func=np.log,
        inverse_func=np.exp
    )
)
```

## Train the Model

```python
# Fit the training data on the regressor by entering it on the pipeline
ml_pipeline.fit(X_train, y_train)
```

## Check Residuals

```python
from yellowbrick.regressor import ResidualsPlot

residuals_plot = ResidualsPlot(ml_pipeline,)
residuals_plot.fit(X_train, y_train)
residuals_plot.score(X_test, y_test)
residuals_plot.show();
```

```python
from yellowbrick.regressor import PredictionError

predictionerror_plot = PredictionError(ml_pipeline)
predictionerror_plot.fit(X_train, y_train)
predictionerror_plot.score(X_test, y_test)
predictionerror_plot.show();
```

The residual plot and prediction error plot above shows that as our predicted value increases, the errors of our model also increases in a linear fashion. If we desire to make accurate inference, the residual plot should exhibit a normally distributed errors around the zero line from the lowest predicted value up-to the highest. Since this is not the case, we cannot make accurate inferences with our linear model.

## Calculate Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

y_train_pred = ml_pipeline.predict(X_train)
y_test_pred = ml_pipeline.predict(X_test)

pd.DataFrame({
    "Metric": ["MAE", "MSE"],
    "Training": [
        mean_absolute_error(y_true=y_train, y_pred=y_train_pred),
        mean_squared_error(y_true=y_train, y_pred=y_train_pred)
    ],
    "Testing": [
        mean_absolute_error(y_true=y_test, y_pred=y_test_pred),
        mean_squared_error(y_true=y_test, y_pred=y_test_pred)
    ]
})
```

The predictions of the model on the training and testing set were off by $\$4,200.00$ on average. It clearly indicates that our model has a bias problem. Since the mean absolute error was close with each other, the model does not exhibit high variance.

## Calculate Feature Importance

```python
fitted_regressor = ml_pipeline.named_steps["transformedtargetregressor"].regressor_

feature_importance = pd.DataFrame(
    data={
        "coefficients": fitted_regressor.coef_
    },
    index=fitted_regressor.feature_names_in_
).sort_values(by="coefficients", ascending=False)
```

```python
feature_importance.plot(kind="bar", title="Linear Regression Feature Importance");
```
